* 필요없는 문자 정제

```python
cleaned_texts = []
for text in data['마케팅']:
    cleaned_text = re.sub(r"[^ㄱ-ㅎㅏ-ㅣ가-힣a-zA-Z0-9\s\[\]]", "", text)
    cleaned_text1 = re.sub(r"담당(.*)자격", "자격", cleaned_text)
    cleaned_text2 = re.sub(r"담당(.*)우대", "우대", cleaned_text1)
    cleaned_text3 = re.sub(r'ㅇ|[\[\]]|\t|\t\t', '', cleaned_text2)
    cleaned_texts.append(cleaned_text3)

clean_mkt = pd.DataFrame({'마케팅': cleaned_texts})

print(clean_mkt)
```

* 채용공고에서 '자격요건' 부분만 추출

```python
extracted_text = []

for text in clean_mkt['마케팅']:
    # 정규식을 사용하여 '자격요건'과 '우대사항' 사이의 내용 추출
    qualification_pattern = re.compile(r'자격(.*)우대|자격(.*)$|요구(.*)\n\n', re.DOTALL)
    qualification_match = qualification_pattern.search(text)

    if qualification_match:
        # 해당하는 그룹을 확인하고 적절한 내용을 추출하여 리스트에 추가합니다.
         if qualification_match.group(1):
             extracted_text.append(qualification_match.group(1))
         elif qualification_match.group(2):
             extracted_text.append(qualification_match.group(2))
         else:
             extracted_text.append(None)
    else:
       extracted_text.append(None)

# 데이터프레임에 새로운 열로 추가합니다.
qualification = pd.DataFrame({'자격요건': extracted_text})

# 결과 확인
print(qualification)
```

```python
qualification.dropna(subset=['자격요건'], inplace=True)
```

```python
qualification['자격요건'] = qualification['자격요건'].apply(lambda x: re.sub(r'요건|사항', '', x))
```

* 채용공고에서 '우대사항' 부분만 추출

```python
extracted_text1 = []

for text in clean_mkt['마케팅']:
    # 정규식을 사용하여 '자격요건'과 '우대사항' 사이의 내용 추출
    qualification_pattern = re.compile(r'우대(.*)$', re.DOTALL)
    qualification_match = qualification_pattern.search(text)

    if qualification_match:
        extracted_text1.append(qualification_match.group(1))
    else:
        extracted_text1.append(None)

# 데이터프레임에 새로운 열로 추가합니다.
qualification['우대사항'] = pd.DataFrame({'우대사항': extracted_text1})

# 결과 확인
print(qualification)
```

* "경험", "어학", "자격증", "전공", "기술", "학력", "업무"로 분류

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")

sequence_to_classify = "Angela Merkel ist eine Politikerin in Deutschland und Vorsitzende der CDU"
candidate_labels = ["경험", "어학", "자격증", "전공", "기술", "학력", "업무"]
# output = classifier(sequence_to_classify, candidate_labels, multi_label=False)
# print(output)
```

* 결과 리스트로 출력

```python
all_results = []

def classify_text(text, candidate_labels):
    output = classifier(text, candidate_labels, multi_label=False)
    result_dict = {
        'sequence': text,
        'label': output['labels'][output['scores'].index(max(output['scores']))],
        'score': max(output['scores']),
    }
    all_results.append(result_dict)
    print(result_dict)

# 'contents' 열의 각 행을 반복하여 결과 출력 및 딕셔너리 저장
for index, row in spec.iterrows():
    sequence_to_classify = row['contents']
    classify_text(sequence_to_classify, candidate_labels)

# 모든 결과를 담고 있는 리스트 출력
print(all_results)
```

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")

sequence_to_classify = ""
candidate_labels = ["영어", "중국어", "일본어", "기타"]
```

```python
all_results = []

def classify_text(text, candidate_labels):
    output = classifier(text, candidate_labels, multi_label=True)
    result_dict = {
        'sequence': text,
        'label': output['labels'][output['scores'].index(max(output['scores']))],
        'score': max(output['scores']),
    }
    all_results.append(output['labels'][output['scores'].index(max(output['scores']))])
    print(result_dict)

# 'contents' 열의 각 행을 반복하여 결과 출력 및 딕셔너리 저장
for index, row in mkt_language.iterrows():
    sequence_to_classify = row['content']
    classify_text(sequence_to_classify, candidate_labels)

# 모든 결과를 담고 있는 리스트 출력
print(all_results)
```
